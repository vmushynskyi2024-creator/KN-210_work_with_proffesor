--- Порівняння dict та класу item 

```python
# --- Варіант із dict ---
item_sword = {
    "name": "Excalibur",
    "rarity": "Legendary",
    "type": "Sword",
    "damage": 50,
    "durability": 100,
}

print("Словник:")
print(item_sword)
print("Тип:", type(item_sword))

# --- Варіант із класом ---

class Item:
    def __init__(self, name, rarity, type_, damage, durability):
        self.name = name
        self.rarity = rarity
        self.type = type_
        self.damage = damage
        self.durability = durability

item_sword_obj = Item("Excalibur", "Legendary", "Sword", 50, 100)

print("\nОб'єкт класу:")
print(item_sword_obj.__dict__)
print("Тип:", type(item_sword_obj))
```

--- Динамічні атрибути у класах

```python
import random

class Item:
    ITEM_MAGIC_TYPES = ["Fire", "Ice", "Lightning", "Earth", "Wind"]

    def __init__(self, name, rarity):
        self.name = name
        self.rarity = rarity
        self.type = "Sword"
        self.damage = 40
        self.durability = 120

item = Item("Dragonblade", "Epic")

item.special_effect = random.choice(["Burn", "Freeze", "Shock"])

print("Атрибути предмету:", item.__dict__)


item2 = Item("Gungnir", "Legendary")
print("Інший предмет:", item2.__dict__)
```

--- Приватні атрибути + property

```python
class Item:
    def __init__(self, name):
        self._name = name   # приватний атрибут
        self.type = "Shield"

    @property
    def name(self):  # доступ тільки для читання
        return f"Назва предмета: {self._name}"

item = Item("Aegis")

print(item.name)


item._name = "Modified Aegis"
print(item.name)
```

--- Методи класу, статичні методи, атака + лікування

```python
import random

class Item:
    def __init__(self, name):
        self._name = name
        self.health = 100
        self.attack_power = 20

    def attack(self, target):
        damage = self.calculate_damage(self.attack_power)
        target.health -= damage
        return f"{self._name} атакує {target._name}! Залишилось HP: {target.health}"

    def heal(self):
        amount = random.randint(5, 15)
        self.health += amount
        return f"{self._name} лікується на {amount}. Тепер HP: {self.health}"

    @staticmethod
    def calculate_damage(base):
        if random.random() < 0.1:
            print("⚡ Критичний удар!")
            return base * 2
        if random.random() < 0.1:
            print("❌ Промах!")
            return 0
        return base

    @staticmethod
    def announce_turn(turn):
        print(f"\n===== Хід {turn} =====")

item1 = Item("Гравець")
item2 = Item("Монстр")

for t in range(1, 4):
    Item.announce_turn(t)

    if random.choice([True, False]):
        print(item1.attack(item2))
    else:
        print(item1.heal())

    if random.choice([True, False]):
        print(item2.attack(item1))
    else:
        print(item2.heal())
```

--- Класовий метод як конструктор

```python
class ItemWithClassMethods:
    def __init__(self, name, attack_power, health):
        self.name = name
        self.attack_power = attack_power
        self.health = health

    @classmethod
    def from_dict(cls, d: dict):
        name = list(d.keys())[0]
        attack, hp = d[name]
        return cls(name, attack, hp)

items_data = {
    "Меч Тіней": [45, 100],
    "Стародавній Щит": [10, 200],
    "Бойовий Лук": [30, 80]
}

print("Створення через звичайний конструктор:")
for name, (a, h) in items_data.items():
    item = ItemWithClassMethods(name, a, h)
    print(item.__dict__)

print("\nСтворення через класовий метод from_dict:")
for name in items_data:
    obj = ItemWithClassMethods.from_dict({name: items_data[name]})
    print(obj.__dict__)
```
