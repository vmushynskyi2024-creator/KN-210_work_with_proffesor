# застосуйте методи класу до всіх обєктів a1-a3

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals}, яку звати {name}, якій {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

    def speak(self):
        if self.animals == "Кіт":
            return f"Моя тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя тваринка {self.animals} говорить Гав"
        else:
            return f"Моя тваринка {self.animals} не розмовляє"

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік"


# Створення об’єктів
a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals("Собака", "Рекс", 2)

# Застосування методів класу до всіх об’єктів
animals_list = [a1, a2, a3]

for animal in animals_list:
    print(animal.speak())
    print(f"{animal.animals} живе зі мною {animal.age} років")
    print(animal.add_one_year())
    print(f"{animal.animals} тепер живе зі мною {animal.age} років")
    print("-" * 40)
```

# задайте декілька обєктів в циклі та виведіть глобальну змінну

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

print(f"В мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")
a1 = MyAnimals("Рибки", "Неони", 1)
MyAnimals("Рибки", "Гупії", 1)
MyAnimals("Рибки", "Соміки", 1)
MyAnimals("Кіт", "Мурка", 5, 10)
print(f"В мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")

print(f"Я доступаюсь до змінної класу з обєкта {a1.total_animal_types}")
a1.total_animal_types = {"Собака"} # я зробив змінну класу атрибутом обєкту
print(f"Я доступаюсь до змінної класу з обєкта {a1.total_animal_types} та з класу {MyAnimals.total_animal_types}")
```

# викличте статичний метод для кожного обєкта a1-a2

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

for animal in [a1, a2]:
    print(f"{animal.animals}: {animal.call_pet()}")

print("\nВиклик статичного методу через клас:")
print(MyAnimals.call_pet())
```

# додайте ще одну properties та виведіть її

```python
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age

    @property
    def info(self):
        return f"Тваринка: {self.animals}, Ім’я: {self.name}, Вік: {self.age} років"

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила зі мною ще один рік"


# Створення об’єкта
a1 = MyAnimals("Кіт", "Мурка", 5)

# Виведення properties
print(a1.info)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")

# Зміна віку
a1.add_one_year()

# Повторне виведення properties
print(a1.info)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
```

# допишіть метод класу який буде превіряти що атрибут name заданий з великої букви

```python
class MyAnimals:
    total_csv = 0
    total_objects = 0

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = int(age)
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        cls._check_name(name)
        return cls(animal, name, age)

    @classmethod
    def _check_name(cls, name: str):
        if not name[0].isupper():
            raise ValueError("Ім’я тваринки повинно починатися з великої літери")

    def show_info(self):
        return f"Мій {self.animals}, звати {self.name}, має {self.age} років"


# Створення об’єктів
a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

# Виведення інформації
print(a1.show_info())
print(a2.show_info())

print(
    f"Загальна кількість об’єктів: {MyAnimals.total_objects}, "
    f"створених з CSV: {MyAnimals.total_csv}"
)
```

# Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```
